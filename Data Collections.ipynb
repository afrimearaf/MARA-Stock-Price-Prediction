```python
import requests
import json
import pandas as pd
from datetime import datetime, timedelta
```

```python
# Set the Tiingo API endpoint URL
url = "https://api.tiingo.com/tiingo/daily/fttusd/prices"

# Set the Tiingo API key
api_key = "692cdb4a9c649324bef05b5bd7f07715cce3d3a7"

# Set the end date for the data you want to retrieve to today's date
end_date = datetime.today().strftime('%Y-%m-%d')

# Set the start date for the data you want to retrieve to 5 years ago from today's date
start_date = (datetime.today() - timedelta(days=365*10)).strftime('%Y-%m-%d')
```

```python
# Set the parameters for the API request
params = {
    "startDate": start_date,
    "endDate": end_date,
    "token": api_key
}
```

```python
# Send the API request and get the response
response = requests.get(url, params=params)

# Parse the JSON data from the response into a Python object
data = json.loads(response.content)
```

```python
# Convert the data to a pandas DataFrame
df = pd.DataFrame(data)

# Print the first 5 rows of the DataFrame
df.head(5)
```

```python
# exports the contents of a pandas DataFrame 'df' to a CSV file named 'MARA_DATA.csv'. 
df.to_csv('fttusd_2024.csv')
```
