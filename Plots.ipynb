```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import mplfinance as mpf
```

```python
# Load the dataset into a DataFrame
df = pd.read_csv('MARA_DATA.csv')
df.head(5)
```

```python
# Convert the 'date' column to datetime if it's not already in that format
df['date'] = pd.to_datetime(df['date'])

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the closing price
ax.plot(df['date'], df['close'], color='blue')

# Set the x-axis label and rotate the ticks for better readability
ax.set_xlabel('Date')
plt.xticks(rotation=45)

# Set the y-axis label
ax.set_ylabel('Closing Price')

# Set the title of the plot
ax.set_title('Closing Price of Stock')

# Display the plot
plt.show()
```

```python
#heatmap of the correlation between variables
corr_matrix = df[['close', 'high', 'low', 'open', 'volume']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['close'], bins=20, kde=True)
plt.title('Closing Price Distribution')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#density plot
plt.figure(figsize=(10, 6))
sns.kdeplot(df['close'], shade=True, color='blue')
plt.title('Closing Price Distribution')
plt.xlabel('Closing Price')
plt.ylabel('Density')
plt.show()
```

```python
#box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='close')
plt.title('Closing Price Distribution')
plt.xlabel('Closing Price')
plt.show()
```

```python
# Set the 'date' column as the index
df.set_index('date', inplace=True)

# Create the candlestick chart
mpf.plot(df, type='candle', title='Stock Price', ylabel='Price')
```

```python
# Plot the area chart
plt.figure(figsize=(10, 6))
plt.fill_between(df.index, df['close'], alpha=0.4)
plt.title('Stock Price')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.show()
```

```python
#Line Chart:
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['close'])
plt.title('Stock Price')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.show()
```

```python
#Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(df.index, df['close'], color='green', label='Closing Price')
plt.bar(df.index, df['open'], color='red', label='Opening Price')
plt.title('Stock Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(df['volume'], df['close'])
plt.title('Stock Price vs. Volume')
plt.xlabel('Volume')
plt.ylabel('Closing Price')
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(10, 6))
plt.violinplot(df['close'], vert=False)
plt.title('Closing Price Distribution')
plt.xlabel('Closing Price')
plt.ylabel('Density')
plt.show()
```

```python
# KDE Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(df['close'], shade=True)
plt.title('Closing Price Distribution (KDE)')
plt.xlabel('Closing Price')
plt.ylabel('Density')
plt.show()
```

```python
# Boxen Plot
plt.figure(figsize=(10, 6))
sns.boxenplot(df['close'])
plt.title('Closing Price Distribution')
plt.xlabel('Closing Price')
plt.show()
```
